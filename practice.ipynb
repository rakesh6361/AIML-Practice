```python
fig, ax1 = plt.subplots(figsize=(8, 5))

color1 = 'tab:blue'
ax1.set_xlabel('Day')
ax1.set_ylabel('Study Hours', color=color1)
ax1.plot(df['Day'], df['Study Hours'], marker='o', linestyle='-', color=color1, label='Study Hours')
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Exam Score', color=color2)
ax2.plot(df['Day'], df['Exam Score'], marker='x', linestyle='--', color=color2, label='Exam Score')
ax2.tick_params(axis='y', labelcolor=color2)

fig.suptitle('Study Hours vs Exam Score Over Days')
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))
plt.grid(True)
plt.show()
```
