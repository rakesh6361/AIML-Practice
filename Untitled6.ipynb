## b) Replace the outliears using mean values & median 

```python
q1 =df["age"].quantile(0.25)

# finding the 3rd quartile
q3 =df['age'].quantile(0.75)

# finding the iqr region
iqr = q3-q1

# finding upper and lower whiskers
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
```

## d)Replacing outliers with median

```python
df = pd.read_csv("titanic.csv")
```

```python
q1 =df["age"].quantile(0.25)

# finding the 3rd quartile
q3 =df['age'].quantile(0.75)

# finding the iqr region
iqr = q3-q1

# finding upper and lower whiskers
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
```

```python
m =df['age'].median()
print(m)
for i in df['age']:
    if i<lower_bound or i>upper_bound :
        df['age'] =df['age'].replace(i,m)
```

## e) Remove data objects with outliers

```python
upperIndex =df[df['age']>upper_bound].index
df.drop(upperIndex,inplace=True)
lowerIndex = df[df['age']<lower_bound].index
df.drop(lowerIndex,inplace=True)
```
